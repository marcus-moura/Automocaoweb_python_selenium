```python
#controla o navegador
from selenium import webdriver
#controla o teclado
from selenium.webdriver.common.keys import Keys
import pandas as pd


#abrir o navegador
nav = webdriver.Chrome()

#Pesquisar no google e pega a cotação do Dolar
nav.get("https://www.google.com/")

#Acessar o campo de busca do google e digitar
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação dolar")

#Clicar no botão pesquisar
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[3]/center/input[1]').send_keys(Keys.ENTER)

#Armazenado o valor da busca do google
cota_dolar = nav.find_element_by_xpath('//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")

#Acessar o campo de busca do google #escrever com o send_keys
nav.get("https://www.google.com/")
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação euro")

#Clicar no botão pesquisar
nav.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

#pegando valor da busca do google
cota_euro = nav.find_element_by_xpath('//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")



#Acessando um site de cotação do ouro
nav.get('https://www.melhorcambio.com/ouro-hoje#:~:text=O%20valor%20do%20grama%20do,em%20R%24%20306%2C10.')
cota_ouro = nav.find_element_by_xpath('//*[@id="comercial"]').get_attribute("value")
#Trocando , por .
cota_ouro = cota_ouro.replace("," , ".")


#Fecha o Chrome
nav.close()
print("-"*60)
print("\t------------Cotações------------")
print(f"Cotação do Dólar: {cota_dolar}")
print(f"Cotação do Euro: {cota_euro}")
print(f"Cotação do Ouro: {cota_ouro}")


#importando a base de dados
bd_produtos = pd.read_excel("Produtos.xlsx")
print("-"*60)
print("\n\t\t\t------------Data Frame------------")
display(bd_produtos)

#passo 1: Atualizar as cotações

bd_produtos.loc[bd_produtos["Moeda"] == "Dólar", "Cotação"] = float(cota_dolar)
bd_produtos.loc[bd_produtos["Moeda"] == "Euro", "Cotação"] = float(cota_euro)
bd_produtos.loc[bd_produtos["Moeda"] == "Ouro", "Cotação"] = float(cota_ouro)

#pass 2: atualizar o preço base reais --> cotação * preço base original
bd_produtos["Preço Base Reais"] = bd_produtos["Cotação"]*bd_produtos["Preço Base Original"]

#passo 3: atualizar o preço final --> Preço base reais * Ajuste
bd_produtos["Preço Final"] = bd_produtos["Preço Base Reais"]*bd_produtos["Ajuste"]

#Ajustando com duas casas decimais
bd_produtos["Preço Final"] = bd_produtos["Preço Final"].map("{:.2f}".format)
bd_produtos["Preço Base Reais"] = bd_produtos["Preço Base Reais"].map("{:.2f}".format)
bd_produtos["Cotação"] = bd_produtos["Cotação"].map("{:.2f}".format)

print("\n\t------------Data Frame com os Dados atualizados------------")
display(bd_produtos)

#Exportando os dados atualizados para um arquivo excel
bd_produtos.to_excel("Preço Atualizado.xlsx") #Para tirar o index da tabela
print("-"*60)
print("\nTabela Exportada com Sucesso!")
```
